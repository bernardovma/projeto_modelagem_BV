```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv', index_col=0)
df.head()
```

```python
# Dados de exemplo
diagnosticos = ['Maligno', 'Benigno']

malignos = df[df['diagnosis'] == 'M']
benignos = df[df['diagnosis'] == 'B']

quantidades = [malignos['diagnosis'].count(), benignos['diagnosis'].count()]

# Cores das fatias do grÃ¡fico de pizza em tons de azul as duas cores
cores = ['#559fcd','#08306b']


plt.figure(figsize=(8, 8))
# Criar o grÃ¡fico de pizza
plt.pie(quantidades, labels=diagnosticos, colors=cores, autopct='%1.1f%%', textprops={'color': 'white'})

# Adicionar tÃ­tulo
plt.title('DiagnÃ³sticos - Maligno vs Benigno', color='white', fontsize=16)

#mudar cor do texto
plt.rcParams['text.color'] = 'white'

# Exibir o grÃ¡fico
plt.show()
```
