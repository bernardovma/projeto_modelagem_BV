```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv', index_col=0)
df.head()
```

```python
#plot de grafico de pizza para cada diagnóstico
df['diagnosis'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(5,5))
plt.title('Diagnóstico')
plt.ylabel('')
plt.show()
```
